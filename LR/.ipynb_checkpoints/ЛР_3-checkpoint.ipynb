# Определение диапазонов варьирования входных проектных параметров

## Исходные данные

```python

p_min = 4 * 1e6

p_max = 20 * 1e6

p_h = 0.1 * 1e6

f_a_max = 0.9

D_n = 200 * 1e-3

F_m = np.pi * D_n ** 2 / 4
hi = 0.98
R_g = 411.52
n_kr = 1.1408
k = n = 1.1597
p_a = 0.1 * 1e6
z = 0.26754
T_r = 3342.4
I_P = 125 * 1e3
t_nom = 5

P = I_P / t_nom

lambda_max = ((n + 1) / (n - 1)) ** 0.5

R_sm = R_g * (1 - z)

A_n = (n_kr * (2 / (n_kr + 1)) ** ((n_kr + 1) / (n_kr - 1))) ** 0.5

betta = (R_sm * hi * T_r) ** 0.5 / A_n

etta_f = P / (p_h * F_m)
# INIT_CONDITIONS = {'p_min': p_min, 'p_max': p_max, 'p_h': p_h, 'f_a_max': f_a_max, 'D_n': D_n,
#                    'F_m': F_m, 'hi': hi, 'R_g': R_g, 'n_kr': n_kr, 'k': k, 'p_a': p_a, 'z': z, 'T_r': T_r, 'I_P': I_P,
#                    't_nom': t_nom, 'P': P, 'lambda_max': lambda_max, 'R_sm': R_sm, 'A_n': A_n, 'betta': betta, 'etta_f': etta_f}
```

## Функции

```python
def lambda_a(p_a, p):
    global lambda_max, n
    return lambda_max * (1 - (p_a / p) ** ((n - 1) / n)) ** 0.5


def dzeta(lam):
    global z, k
    return (1 - z) + z * (2 * k) / (k + 1) * (lam) ** 2 / (lam ** 2 + 1)


def eps(lam):
    global k
    return (1 - (k - 1) / (k + 1) * lam ** 2) ** (1 / (k - 1))


def pi(lam):
    global k
    return (1 - (k - 1) / (k + 1) * lam ** 2) ** (k / (k - 1))


def q(lam):
    global k
    return lam * eps(lam) / ((2 / (k + 1)) ** (1 / (k - 1)))


def tau(lam):
    global k
    return 1 - ((k - 1) / (k + 1)) * lam ** 2


def I_ud(p, lam):
    global betta, p_h
    return betta * ((lam + lam ** -1) * eps(1) * dzeta(lam) - p_h / p * 1 / q(lam))


def f_a_bound1(p, lam):
    global etta_f, p_h
    return etta_f / (p / p_h * (lam ** 2 + 1) * eps(lam) * dzeta(lam) - 1)


def f_a_bound2(p, lam):
    global etta_f, p_h
    return etta_f / (p / p_h * (lam ** 2 + 1) / tau(lam) * dzeta(lam) - 1)


def f_a_bound3(I_ud3, lam):
    global etta_f, betta
    return etta_f * (betta / I_ud3 * (lam + lam ** -1) * eps(1) * dzeta(lam) - 1)


def v_a(lam):
    return 1 / q(lam)


def p_a3(p, lam):
    return p * pi(lam)
def u(p):
    return u_1 * (p) ** nu

def fi_kappa(kappa):
    if kappa > kappa_por:
        return 1 + 0.003 * (kappa - kappa_por)
    else:
        return 1
```

## Первая точка

```python
lambda_a1 = lambda_a(p_a, p_min)
f_a1 = f_a_bound1(p_min, lambda_a1)
v_a1 = v_a(lambda_a1)
I_ud1 = I_ud(p_min, lambda_a1)
print('lambda_a1 =', lambda_a1)
print('v_a1 =', v_a1)
print('f_a1 =', f_a1)
print('p_nom1 =', p_min)
print('p_a1 / p_h1 =', p_a3(p_min, lambda_a1) / p_h)
print('I_ud1 =', I_ud1)
```

## Вторая точка

```python
lambda_a2 = lambda_a(p_a, p_max)
f_a2 = round(f_a_bound1(p_max, lambda_a2), 2)
v_a2 = v_a(lambda_a2)
I_ud2 = I_ud(p_max, lambda_a2)
print('lambda_a2 =', lambda_a2)
print('v_a2 =', v_a2)
print('f_a2 =', f_a2)
print('p_nom2 =', p_max)
print('p_a2 / p_h2 =', p_a3(p_max, lambda_a2) / p_h)
print('I_ud2 =', I_ud2)
```

## Третья точка

```python
N = 1000
i = 1
while i <= lambda_a1 - 0.1:
    if I_ud(p_max, i) < I_ud1:
        I_ud3 = I_ud(p_max, i)
        lambda_a3 = i
    else:
        break
    i += 0.001
f_a3 = f_a_bound3(I_ud3, lambda_a3)
v_a3 = v_a(lambda_a3)
print('lambda_a3 =', lambda_a3)
print('v_a3 =', v_a3)
print('f_a3 =', f_a3)
print('p_nom3 =', p_max)
print('p_a3 / p_h3 =', p_a3(p_max, lambda_a3) / p_h)
print('I_ud3 =', I_ud3)
```

# Обоснование проектных параметров РДТТ

## Определение диапазонов варьирования

```python
f_a_range = np.linspace(f_a3, f_a1, 50)
l_a_range = np.linspace(lambda_a3, lambda_a2, 50)
m_dv_arr = np.array([])
l_dv_arr = np.array([])
f_a_arr = np.array([])
v_a_arr = np.array([])
f_a_false = np.array([])
v_a_false = np.array([])

D_ks = 0.96 * D_n  # m # Change
q_i = [0.137, 0.71, 0.15,0.003]  # change
ro_i = [920, 1950, 2700, 5250]
total = 0

for i in range(len(ro_i)):
    total += q_i[i] / ro_i[i]

ro_t = round(1 / total)

F_ks = np.pi * (D_ks) ** 2 / 4  # m**2

b = D_n / 4
count = 4
a_ = 0.3
c_ = 0.05
delta_p1, delta_p2 = 6, 6
delta_zks = 1
delta_vks = 15
delta_ps1 = 6
delta_ps2 = 3

tetta_1 = np.radians(30)
tetta_2 = np.radians(15)

ro_p = 1500
ro_kor = 7830
ro_zks = 920
ro_ps = 1750
ro_vks = 2200

ro_cil = np.array([ro_kor, ro_zks])
ro_dn_1 = np.array([ro_kor, ro_p])
ro_dn_2 = np.array([ro_kor, ro_p])
ro_st = np.array([ro_kor, ro_vks])
ro_kon_1 = np.array([ro_kor, ro_p])
ro_kon_2 = np.array([ro_kor, ro_ps])
ro = np.array([ro_cil, ro_dn_1, ro_dn_2,
              ro_st, ro_kon_1, ro_kon_2])
sigma_vr = 1700
sigma_02 = 1400

k_1 = 1.1
k_2 = 1.2
etta = 1.25

D_t = 0.0012
nu = 0.358
u_1 = 6.4 * 1e-3
T_0 = 273.15+50
T_ref = 293.15
kappa_por = 100



for f in f_a_range:
    for l in l_a_range:
        I_udp = betta * (l + l ** -1) * eps(1) * dzeta(l)
        I_ud = I_udp * etta_f / (etta_f + f)
        p_nom = p_h * betta * v_a(l) / (I_udp - I_ud) / 1e6
        p_a = p_a3(p_nom, l)
        e_d = 2 / D_ks * u(p_nom) * t_nom
        if e_d >= 1 - c_ and (p_a > p_h*1e-6):
            f_a_false = np.append(f_a_false, f)
            v_a_false = np.append(v_a_false, v_a(l))
            break
        else:
            d_ = 1 - e_d
            eps_f = 1 - d_ ** 2
            f_sh = count / np.pi * (c_ * (1 - c_ ** 2)**0.5 - c_ *
                                    (d_**2-c_**2)**0.5 + np.arcsin(c_) - d_**2 * np.arcsin(c_ / d_))
            eps_omega = eps_f - a_ * f_sh
            omega = I_P / I_ud
            l_zar = omega / (ro_t * eps_omega * F_ks)
            kappa = 4 * l_zar / (D_ks * (1 - e_d))
            p_50 = p_nom * np.exp((D_t * (T_0 - T_ref)) /
                                  (1-nu)) * fi_kappa(kappa) ** (1 / (1-nu))
            p_p = p_50 * sigma_vr / sigma_02 * k_1 * k_2 * etta

            delta_ob = D_n / 2 * p_p / sigma_vr
            delta_dn = D_n / 2 * p_p / sigma_vr * ((D_n / b)**2/24 + 1 / 3)
            D_a = np.sqrt(4 / np.pi * f * F_m)

            D_kr = D_a / np.sqrt(v_a(l))

            l_ob = l_zar
            delta_zks = 1*1e-3
            b_1 = b
            b_2 = b
            delta_dn1 = delta_dn
            delta_dn2 = delta_dn
            delta_p1 = 6*1e-3
            delta_p2 = 6*1e-3
            d = D_n / 2
            D_vh = D_n / 2
            D_vih = D_a
            tetta_1 = np.radians(30)
            tetta_2 = np.radians(15)
            l_kr = D_kr / 2
            delta_s1 = 2 * delta_ob
            delta_s2 = delta_ob
            delta_st = 3 * delta_ob
            delta_ps1 = 6*1e-3
            delta_ps2 = 3*1e-3
            delta_vks = 15*1e-3
            INIT_V = [D_n, l_ob, delta_ob, delta_zks, b_1, b_2, delta_dn1, delta_dn2, delta_p1, delta_p2, d, D_vh,
                      D_vih, D_kr, tetta_1, tetta_2, l_kr, delta_s1, delta_s2, delta_st, delta_ps1, delta_ps2, delta_vks]
            V_cil = Volume(INIT_V).cil_ks()
            V_dn_1 = Volume(INIT_V).ell_pd()
            V_dn_2 = Volume(INIT_V).ell_sd()
            V_st = Volume(INIT_V).cil_st()
            V_kon_1 = Volume(INIT_V).kon_1()
            V_kon_2 = Volume(INIT_V).kon_2()
        
            V = np.array([V_cil, V_dn_1, V_dn_2, V_st, V_kon_1, V_kon_2])
            m_dv_i = ro * V  # true
            m_dv_zer = [[0] for i in range(len(V))]

            for i in range(len(V)):
                m_dv_zer[i] = m_dv_i[i][0] + m_dv_i[i][1]
            m_dv_0 = sum(m_dv_zer)

            m_dv = m_dv_0 + omega

            alpha_dv = m_dv_0 / omega
            l_dv = b + l_ob + b * np.sqrt(1 - (D_vh / D_n)**2) + (D_vh - D_kr) / \
                (2 * np.tan(tetta_1)) + D_kr / 2 + \
                (D_a - D_kr) / (2 * np.tan(tetta_2))
            if (p_a > p_h*1e-6) and (p_nom < p_max*1e-6) and (I_ud > I_ud1) and (m_dv > 0) and (kappa < 200):
                f_a_arr = np.append(f_a_arr, f)
                v_a_arr = np.append(v_a_arr, v_a(l))
                m_dv_arr = np.append(m_dv_arr, m_dv)
                l_dv_arr = np.append(l_dv_arr, l_dv)
            elif (kappa >= 200) and (p_a > p_h*1e-6):
                f_a_false = np.append(f_a_false, f)
                v_a_false = np.append(v_a_false, v_a(l))
c_dv_arr = np.sqrt((m_dv_arr * l_dv_arr) / (min(m_dv_arr) * min(l_dv_arr)))
c_dv_opt = min(c_dv_arr)
print(c_dv_opt)
print(m_dv_arr[c_dv_arr == min(c_dv_arr)])
print(l_dv_arr[c_dv_arr == min(c_dv_arr)])
```
