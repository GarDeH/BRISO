```python
def S_g_B(e):
    def S1(e):
        return 4*np.pi*(R_v-e)*h1(e) + 2*np.pi*(r_v - e)*c1(e)

    def h1(e):
        return (R_v- e) - np.sqrt((R_v - e)**2 - (r_v - e)**2)

    def c1(e):
        return 2*(e_B_0 - e - h1(e))
    
    def S2(e):
        return 4*np.pi*(R_v - e)*(e_B_0 - e)

    def e1(e):
        return ((R_v**2 - r_v**2) - (R_v - e_B_0)**2)/(2*(R_v - r_v))
    ee = e1(e)
    
    def WW(e):
        return np.pi*r_v**2*c_v + 2*np.pi*h_v**2*(R_v - h_v/3)
    if e < ee:
        return S1(e)
    else:
        return S2(e)


def X(G_B, T, G_t, dQ_t_dt, F_q_1_):
    x = G_B*c_p_0*(T_p_0-T)+G_t*c_p_1*(T_p_1-T)-dQ_t_dt*F_q_1_  # c_p_1
    return x


def Y(G_B, T, G_t, p, rho, G):
    y = G_B*R_0*T+G_t*R_1*T-p/rho*G-p*(G_B/rho_B + G_t/rho_t)
    return y

# закон горения
def U_t(p):
    return u_1_t*(p/p_ref_t)**nu_t*np.exp(D_t_t*(T_0-T_ref))


def U_B(p):
    return u_1_B*(p/p_ref_B)**nu_B*np.exp(D_t_B*(T_0-T_ref))


def FI_kappa(kappa):  # check
    if kappa < kappa_ref:
        return 1
    else:
        return 1 + 0.003*(kappa-kappa_ref)

def F_q_1(etta_T):
    if 0 < etta_T_s - etta_T:
        return 1
    else:
        return 0


def F_0(e):
    if  0 <= e_B_0 - e:
        return 1
    else:
        return 0


def F_1(etta_T, e):
    if 0 <= z_0 - e:
        return 1 - F_q_1(etta_T)
    else:
        return 0

# расчёт теплового потока


def q(etta_T, G_B, T):
    Pr = (c_p_g*mu_g)/lamda_g
    Re = G_B/F_sv*d/mu_g
    Nu = 0.023*Re**0.8*Pr**0.4
    alpha = Nu*lamda_g/d
    return alpha*(T-(T_0+etta_T**0.5))


def Pi(lamda):
    return (1 - (n_t-1)/(n_t+1)*lamda**2)**(n_t/(n_t-1))


def G_(p, R, T):  # секундный массовый расход ПС через сопло
    if p*Pi(1) < p_h:
        return ((p*F_kr)/(np.sqrt(R*T)))*(p_h/p)**(1/n_kr)*np.sqrt(2*n_kr/(n_kr-1)*np.abs(1-(p_h/p)**((n_kr-1)/n_kr)))
    else:
        return (A_n*F_kr*p)/((R*T)**0.5)
    

def G_b(S, u, F_0):  # расход от воспламенителя
    return S*u*rho_B*F_0


def G_topl(S, u, fi_kappa, f_1):  # расход от основного заряда
    return S*u*rho_t*f_1*fi_kappa


def S_g_t(z):  # зависимость площади поверхности горения основного заряда
    def S_a(z):
        return np.pi*(d + 2*z)*(l_gr - a - z)


    def S_b(z):
        if z > z_1:
            return 0
        elif z <= z_1:
            return n*(beta - fi_1(z))*(d + 2*z)*a

    def fi_1(z):
        return np.arcsin((c + 2*z)/(d + 2*z))

    def S_c(z):
        if 0 <= z <= z_1:
            return 2*n*a*b1(z)
        elif z_1 < z <= z_2:
            return 2*n*a*b2(z)
        elif z > z_2:
            return 0

    def get_kappa(z):
        if z <= z_1:
            return S_c(z) / (F * f_cut)
        elif z <= z_2:
            return (S_a(z) + S_b(z)) / (F * (1 - epsilon_f))
        
    def b1(z):
        return 0.5*(np.sqrt(D**2 - (c + 2*z)**2) - np.sqrt((d + 2*z)**2 - (c + 2*z)**2))

    def b2(z):
        return 0.5*(np.sqrt(D**2 - (c + z*2)**2) - (c + 2*z)*np.tan(beta)**(-1))

    def S_d(z):
        return np.pi/4*(D**2 - (d + 2*z)**2)


    def S(z):
#         print(S_a(z),S_b(z),S_c(z),S_d(z))
        return S_a(z) + S_b(z) + S_c(z) + S_d(z)
    
    z_1 = 0.5*(d*np.sin(beta)-c)/(1-np.sin(beta))
    z_2 = (D*np.sin(beta)-c)/2
    return S(z), get_kappa(z)
        

def rk4(init, sys, t0, dt, n, stop):

    # declaration

    t = np.zeros(n)
    m = len(init) + 1
    res = np.zeros((n, m))

    # initialization

    i = 0
    X = np.array([i for i in init])
    t[i] = t0
    res[i, 0] = t[i]
    res[i, 1:m] = X

    # main loop
    # while stop(t[i], X) > 0 and i < n - 1:
    while stop(X[0]) and i < n - 1:
        k1 = sys(t[i], X)
        k2 = sys(t[i] + 0.5*dt, X + k1*0.5*dt)
        k3 = sys(t[i] + 0.5*dt, X + k2*0.5*dt)
        k4 = sys(t[i] + dt, X + k3*dt)
        i += 1
        X += (k1 + 2*k2 + 2*k3 + k4)*dt/6
        t[i] = t[i - 1] + dt
        res[i, 0] = t[i]
        res[i, 1:m] = X
    return res[0:i + 1, :]



def System(t, y):
    p, T, W, c_p, R, e_B, e, etta_T = y
    k = c_p/(c_p-R)
    rho = p/(R*T)
    f_q_1 = F_q_1(etta_T)
    f_0 = F_0(e_B)
    f_1 = F_1(etta_T, e)
    u_B = U_B(p)
    u = U_t(p)
    S_B = N*S_g_B(e_B)
    S_t = S_g_t(e)[0]

    
    kappa = S_g_t(e)[1]
    fi_kappa = FI_kappa(kappa)
    
    G_B = G_b(N*S_g_B(e_B), u_B, f_0)
    G_t = G_topl(S_g_t(e)[0], u, fi_kappa, f_1)
    G = G_(p, R, T)
    

    q_1 = q(etta_T, G_B, T)
    dQ_t_dt = q_1*S_0

    x = X(G_B, T, G_t, dQ_t_dt, f_q_1)
    y_ = Y(G_B, T, G_t, p, rho, G)
    print(S_t, kappa, fi_kappa, G_B, G_t, G,q_1,x, y_)
    dp_dt = (k-1)/W*(x+y_*k/(k-1))
    dT_dt = (k - 1)/(rho*W*R)*(x+y_)  # 1/
    dW_dt = G_B/rho_B+G_t/rho_t
    dc_p_dt = 1/(rho*W)*(G_B*(c_p_0-c_p)+G_t*(c_p_1-c_p))
    dR_dt = 1/(rho*W)*(G_B*(R_0-R)+G_t*(R_1-R))
    de_b_dt = u_B*f_0
    de_dt = u*fi_kappa*f_1
    detta_T_dt = (2*q_1**2)/(c_t*lamda_t*rho_t)*f_q_1

    return np.array([dp_dt, dT_dt, dW_dt, dc_p_dt, dR_dt, de_b_dt, de_dt, detta_T_dt])


def f_res(INIT, t_k_0, t_k_1, d_t, stop):
    Res = rk4(INIT, System, t_k_0, d_t, round(t_k_1/d_t), stop)
    print(d_t)
    e_topl = Res[:, 7]
    p_kam = Res[:, 1]
    e_vospl = Res[:, 6]
    t_kam = Res[:, 0]
    T_kam = Res[:, 2]
    return [Res, {'e_topl': e_topl, 'p_kam': p_kam, 'e_vospl': e_vospl, 't_kam': t_kam, 'T_kam': T_kam}]
```
